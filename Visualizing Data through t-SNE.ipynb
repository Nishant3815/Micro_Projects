```python
import numpy as np 
from sklearn.datasets import fetch_mldata
```

```python
mnist = fetch_mldata("MNIST original")
X = mnist.data/255
y = mnist.target
```

```python
mnist.data
```

```python
print(X.shape,y.shape)
```

```python
import pandas as pd
feat_cols = ['pixel'+str(i) for i in range(X.shape[1])]
df = pd.DataFrame(X,columns=feat_cols)
df['label']=y
df['label'].apply(lambda i : str(i))
X,y = None,None
print(df.shape)
```

```python
pd.set_option('display.max_columns', 500)
df.head()
```

```python
#Since the dataset is sorted we need to take the random permutations
rndperm = np.random.permutation(df.shape[0])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rndperm[8]
```

```python
# Plot the graph
plt.gray()
fig = plt.figure( figsize=(10,10) )
for i in range(0,24):
    ax = fig.add_subplot(6,4,i+1, title='Digit: ' + str(df.loc[rndperm[i],'label']) )
    ax.matshow(df.loc[rndperm[i],feat_cols].values.reshape((28,28)).astype(float))
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df[feat_cols].values)
df['pca_one'] = pca_result[:,0]
df['pca_two'] = pca_result[:,1]
df['pca_three'] = pca_result[:,2]
print(pca.explained_variance_ratio_)
```

```python
df.head()
```

```python
from ggplot import *
chart = ggplot(df.loc[rndperm[:3000],:],aes(x='pca_one',y='pca_two',color = 'label'))\
           + geom_point(size=100,alpha=0.8)\
    + ggtitle("First and Second Principal Components colored by digit")
chart
```

```python
import time 
from sklearn.manifold import TSNE
n_sne = 6000
time_start = time.time()
tsne = TSNE(n_components=2,verbose=1,perplexity=40,n_iter=300)
tsne_results = tsne.fit_transform(df.loc[rndperm[:n_sne],feat_cols].values)
```

```python
print('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
df_tsne = df.loc[rndperm[:n_sne],:].copy()
df_tsne['x_tsne'] = tsne_results[:,0]
df_tsne['y_tsne'] = tsne_results[:,1]
chart = ggplot( df_tsne, aes(x='x_tsne', y='y_tsne', color='label') ) \
             + geom_point(size=70,alpha=0.1) \
    + scale_color_brewer(palette='spectral',type='div')\
    + ggtitle("tSNE dimensions colored by digit") \
    
chart
```
